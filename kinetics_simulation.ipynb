```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from datetime import datetime
from sklearn.linear_model import LinearRegression
print('libraries imported')
```

```python
## read INP data into a dataframe. 'days' indicate the day of August
fname = '/private/icg140/SAPHIR2018/FM_PTR_INP.csv'
df = pd.read_csv(fname, sep=',', parse_dates=["time"])
df.columns = ['Index', 'Datetime', 'PTR_INP (ppb)', 'FM_INP (norm counts)'] 
df['days'] = pd.DatetimeIndex(df['Datetime']).day
df.head()
```

```python
### Display all data together

humid = df[(df['days'] >= 10)]

sns.set()
humid.dropna(axis=0, inplace=True)
x = humid['FM_INP (norm counts)'].values.reshape(-1,1)
y = humid['PTR_INP (ppb)'].values.reshape(-1,1)
lmreg = LinearRegression() 
lmreg.fit(x, y)
print(lmreg.intercept_)
print(lmreg.coef_)
g = sns.jointplot(x='FM_INP (norm counts)', y='PTR_INP (ppb)', data=humid, kind="reg", color="m")
plt.show()
```

```python
### Display all data from the aerosol phase

humid = df[(df['days'] >= 14)]

sns.set()
humid.dropna(axis=0, inplace=True)
x = humid['FM_INP (norm counts)'].values.reshape(-1,1)
y = humid['PTR_INP (ppb)'].values.reshape(-1,1)
lmreg = LinearRegression() 
lmreg.fit(x, y)
print(lmreg.intercept_)
print(lmreg.coef_)
g = sns.jointplot(x='FM_INP (norm counts)', y='PTR_INP (ppb)', data=humid, kind="reg", color="m")
plt.show()
```

```python
import netCDF4
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from netCDF4 import num2date
```

```python
def get_smps(path):
    with netCDF4.Dataset(path, 'r') as nc:

        nctime = nc.variables['time'][:]
        t_unit = nc.variables['time'].units
        dt = num2date(nctime, units=t_unit, calendar='standard')
        df = pd.DataFrame(data={'time': dt})
        
        keys = nc.variables.keys()
        smps = []
        for item in keys:
            if 'SMPS' in item and '@STDEV' not in item:
                df[item] = nc.variables[item]
        df.dropna(inplace=True)
        df.set_index('time', inplace=True)
        return df
```
